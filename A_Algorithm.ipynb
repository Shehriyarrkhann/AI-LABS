# A* Algorithm Implementation





```python
import heapq

# Graph (adjacency list with weights)
graph = {
    'S': {'A': 1, 'B': 4},
    'A': {'C': 2, 'D': 5},
    'B': {'D': 1, 'E': 7},
    'C': {'G': 5},
    'D': {'G': 3},
    'E': {'G': 2},
    'G': {}
}

# Heuristic values (estimated cost to goal G)
heuristic = {
    'S': 7,
    'A': 6,
    'B': 5,
    'C': 3,
    'D': 2,
    'E': 4,
    'G': 0
}

def a_star(graph, start, goal):
    open_list = [(heuristic[start], 0, start, None)]

    came_from = {}
    g_cost = {start: 0}

    while open_list:
        f, g, current, parent = heapq.heappop(open_list)

        if current in came_from:
            continue

        came_from[current] = parent

        if current == goal:
            break

        for neighbor, weight in graph[current].items():
            new_g = g + weight

            if neighbor not in g_cost or new_g < g_cost[neighbor]:
                g_cost[neighbor] = new_g
                f_score = new_g + heuristic[neighbor]
                heapq.heappush(open_list, (f_score, new_g, neighbor, current))

    # Reconstruct path
    path = []
    node = goal
    while node is not None:
        path.append(node)
        node = came_from.get(node)

    path.reverse()

    return path, g_cost.get(goal, float('inf'))

start = 'S'
goal = 'G'

path, cost = a_star(graph, start, goal)

print("Shortest Path:", path)
print("Total Cost:", cost)
```
