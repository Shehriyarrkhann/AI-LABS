# **QUESTION 02**

```python
import math

# Minimax function
def minimax(depth, nodeIndex, isMax, values, maxDepth):

    # Base case: leaf node
    if depth == maxDepth:
        return values[nodeIndex]

    if isMax:
        return max(
            minimax(depth + 1, nodeIndex * 2, False, values, maxDepth),
            minimax(depth + 1, nodeIndex * 2 + 1, False, values, maxDepth)
        )
    else:
        return min(
            minimax(depth + 1, nodeIndex * 2, True, values, maxDepth),
            minimax(depth + 1, nodeIndex * 2 + 1, True, values, maxDepth)
        )


# Example leaf nodes (from your board)
values = [-4, -5, 2, 3, 4, 5, 6, 7]

# Tree depth = log2(n)
depth = int(math.log2(len(values)))

result = minimax(0, 0, True, values, depth)

print("Optimal Value (Minimax):", result)
```

# **QUESTION 01**

```python
def minimax(node, is_max, graph, values):

    # If leaf node → return its value
    if node not in graph:
        return values[node]

    children = graph[node]

    if is_max:
        best = float("-inf")
        for child in children:
            best = max(best, minimax(child, False, graph, values))
        return best

    else:
        best = float("inf")
        for child in children:
            best = min(best, minimax(child, True, graph, values))
        return best


# ---------------- GRAPH (FULL 3 LEVEL TREE) ----------------
graph = {
    "A": ["B", "C"],

    "B": ["D", "E"],
    "C": ["F", "G"],

    "D": ["H", "I"],
    "E": ["J", "K"],
    "F": ["L", "M"],
    "G": ["N", "O"]
}

# ---------------- LEAF VALUES (LEVEL 3) ----------------
values = {
    "H": 3,
    "I": 5,
    "J": -2,
    "K": 4,
    "L": 6,
    "M": 1,
    "N": -3,
    "O": 2
}

# ---------------- RUN MINIMAX ----------------
result = minimax("A", True, graph, values)

print("Result:", result)
```

```python
import math

def alpha_beta(node, depth, alpha, beta, is_maximizing, tree, leaf_values):
    # Base case: leaf node
    if node not in tree:
        return leaf_values[node]

    children = tree[node]

    if is_maximizing:
        max_eval = -math.inf
        for child in children:
            eval = alpha_beta(child, depth + 1, alpha, beta, False, tree, leaf_values)
            max_eval = max(max_eval, eval)
            alpha = max(alpha, eval)
            if beta <= alpha:
                print(f"  Pruning at node {child} (β={beta} ≤ α={alpha})")
                break  # β cut-off
        return max_eval
    else:
        min_eval = math.inf
        for child in children:
            eval = alpha_beta(child, depth + 1, alpha, beta, True, tree, leaf_values)
            min_eval = min(min_eval, eval)
            beta = min(beta, eval)
            if beta <= alpha:
                print(f"  Pruning at node {child} (β={beta} ≤ α={alpha})")
                break  # α cut-off
        return min_eval


# Tree structure (adjacency list)
tree = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F', 'G'],
    'D': ['n1', 'n2'],
    'E': ['n3', 'n4'],
    'F': ['n5', 'n6'],
    'G': ['n7', 'n8'],
}

# Leaf node values
leaf_values = {
    'n1': 4, 'n2': 3,
    'n3': 2, 'n4': 10,
    'n5': 1, 'n6': 4,
    'n7': 4, 'n8': 5
}

result = alpha_beta('A', 0, -math.inf, math.inf, True, tree, leaf_values)
print(f"\nOptimal Value: {result}")
```
