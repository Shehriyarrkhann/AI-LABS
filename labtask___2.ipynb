# Question 01

```python
# Class for Trapezoid
class Trapezoid:
    def __init__(self, a, b, h):
        self.a = a
        self.b = b
        self.h = h

    def area(self):
        return (self.a + self.b) * self.h / 2


# Class for Parallelogram
class Parallelogram:
    def __init__(self, base, height):
        self.base = base
        self.height = height

    def area(self):
        return self.base * self.height


# Different sets of dimensions
t1 = Trapezoid(10, 8, 5)
t2 = Trapezoid(6, 4, 3)

p1 = Parallelogram(12, 6)
p2 = Parallelogram(10, 5)


# Store areas in dictionary
areas = {}

areas["Trapezoid1"] = t1.area()
areas["Trapezoid2"] = t2.area()
areas["Parallelogram1"] = p1.area()
areas["Parallelogram2"] = p2.area()


# Display all areas
print("Areas of shapes:")
for shape in areas:
    print(shape, "=", areas[shape])


# Find shape with maximum area
max_shape = max(areas, key=areas.get)

print("\nShape with maximum area:", max_shape)
print("Maximum area:", areas[max_shape])
```
